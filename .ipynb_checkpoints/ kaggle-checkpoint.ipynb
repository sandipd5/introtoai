```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from lightgbm import LGBMRegressor
# from catboost import CatBoostRegressor
from sklearn.svm import SVR
```

```python
# constants
columns_to_drop = ['type','packageType', 'trip','exWeatherTag']
columns_to_encode = ['dayPart','carrier','originLocation','destinationLocation']
prediction_variable = 'cost'
models = ['XGB', 'Decision_Tree_Regressor', 'Random_Forest', 'Gradient_Boosting', 'LightGBM', 'CatBoost', 'SVM']
model_to_use = models[1]

# function

# function to apply one-hot-encoding
def one_hot(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode 
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df.drop(each, axis=1), dummies], axis=1)
    return df


# function to drop column from given dataset
def dropSelectedColumns(df, columns):
    count = 0
    for column in columns:
        print('------------------------------ ')
        print(f'Column to drop : {column}')
        if (column != 'trip'):
            print('Unique Values in the column : ')
            print(df[column].unique())
            print('Value count of the column : ')
            print(df[column].value_counts())
        df = df.drop(column, axis = 1)
        count += 1
        print(f"Total Dropped Columns {count}")
        print('------------------------------ \n')
    print(f'Dataset after drop : ')
    print(df.head(10))
    return df

# use days since reference using currentdate and drop 
def convert_date_to_days_since(df):
    df['date'] = pd.to_datetime(df['date'] )
    current_date = pd.to_datetime(datetime.now().strftime('%Y-%m-%d'))
    df['days_since_reference'] = (current_date-df['date']).dt.days
    df = df.drop('date', axis=1)
    return df
```

```python
# Prompt the user to choose a model
print("Available models:")
for i, model_name in enumerate(models):
    print(f"{i}: {model_name}")

user_choice = input("Enter the number corresponding to the model you want to use: ")
user_choice = int(user_choice)

if user_choice not in range(len(models)):
    print("Invalid choice. Please choose a number between 0 and", len(models)-1)
else:
    model_to_use = models[user_choice]
```

```python
print(df)
df_encoded = one_hot(df,columns_to_encode)
print(df_encoded)
```

```python
X = df_encoded.drop(prediction_variable, axis=1)
y = df_encoded[prediction_variable]
print("Independent variables \n",X)
print("Dependent Variable \n",y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.01, random_state = 42)
print('X_train : ', X_train)
print('X_test : ', X_test)
print('y_train : ', y_train)
print('y_test : ', y_test)
```

```python
model = None
print(f'Model being used {model_to_use}')
if model_to_use == 'XGB':
    model = XGBRegressor()
elif model_to_use == 'Decision_Tree_Regressor':
    model = DecisionTreeRegressor()
elif model_to_use == 'Random_Forest':
    model = RandomForestRegressor()
elif model_to_use == 'Gradient_Boosting':
    model = GradientBoostingRegressor()
elif model_to_use == 'LightGBM':
    model = LGBMRegressor()
elif model_to_use == 'CatBoost':
    model = CatBoostRegressor()
elif model_to_use == 'SVM':
    model = SVR()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Mean absolute error : ", mean_absolute_error(y_test, y_pred))
print("Mean square error : ", mean_squared_error(y_test, y_pred))
```

```python
# Loading test dataset to perform prediction
df_test = pd.read_csv('datasets/test.csv')
```

```python
# Convert 'date' to numerical feature (e.g., number of days since a reference date)
df_test = convert_date_to_days_since(df_test)

df_test_preprocessed = dropSelectedColumns(df_test, columns_to_drop)

df_test_preprocessed = one_hot(df_test_preprocessed, columns_to_encode)
df_test_preprocessed.head(10)
```

```python
test_pred = model.predict(df_test_preprocessed)
print(test_pred)
```

```python
df_output = pd.DataFrame()
df_output['trip'] = df_test['trip']
df_output['cost'] = test_pred
df_output.to_csv('Cost Prediction/Output_files/'+model_to_use + '_prediction.csv',index=False)
```
